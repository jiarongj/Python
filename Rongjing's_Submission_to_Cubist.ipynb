```python
import matplotlib.pyplot as plt
import yfinance as yf

%config InlineBackend.figure_format = 'svg'
```

```python
tickers = ['MSFT', 'ORCL', 'IBM', 'AAPL', 'BMRC']
start_date = '2018-10-01'
end_date = '2018-12-31'
data = yf.download(tickers = tickers, start = start_date, end = end_date)
```

```python
data = yf.download(tickers, start_date, end_date)
data.head()
```

```python
data.columns
```

```python
df = data['Adj Close']
df.head()
```

```python
df.plot()
```

```python
w = [32, 100, 150,350, 210]
res_data = df.pct_change()[1:]
res_data
```

```python
weighted_returns = (w * res_data)
weighted_returns
```

```python
portf_returns = weighted_returns.sum(axis=1)
portf_returns
```

```python
fig = plt.figure()
ax1 = fig.add_axes([0.1,0.1,0.8,0.8])
ax1.hist(portf_returns, bins = 60)
ax1.set_xlabel('Portfolio returns')
ax1.set_ylabel("Freq")
ax1.set_title("Portfolio Returns by days")
plt.show(); 
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('/content/file_destination/stocks_latest/earnings_latest.csv')
new_df = df[df['symbol'].isin(['AAPL','IBM','MSFT','ORCL','BMRC'])][df['qtr'].isin(['04/2019','03/2019','02/2019','01/2019','12/2018','11/2018','10/2018','09/2018','08/2018'])]
new_df
```

```python
prep = new_df[['symbol', 'qtr','eps_est']]
prep
plt.style.use('seaborn-whitegrid')
aapl,bmrc,msft,ibm,orcl = prep[prep['symbol']=='AAPL'],prep[prep['symbol']=='BMRC'],prep[prep['symbol']=='IBM'],prep[prep['symbol']=='MSFT'],prep[prep['symbol']=='ORCL']
fig, axes = plt.subplots(nrows=3, ncols=2, figsize=(22,18))
ax11 = axes[0][0]
ax12 = axes[0][1]
ax21 = axes[1][0]
ax22 = axes[1][1]
ax31 = axes[2][0]
line_width = 2.5
lower = prep['eps_est'].min() - 0.3
higher = prep['eps_est'].max() + 0.3

index = [0,1,2]
aapl['index'], bmrc['index'], msft['index'], ibm['index'], orcl['index'] = index, index, index, index, index[:2]
aapl, bmrc, msft, ibm, orcl = aapl.set_index('index'), bmrc.set_index('index'), msft.set_index('index'), ibm.set_index('index'), orcl.set_index('index')

ax3 = aapl.plot(x='qtr', y='eps_est', ax=ax11, c='blue', legend=False, lw=line_width, ylim=(lower, higher));
bx3 = bmrc.plot(x='qtr', y='eps_est', ax=ax12, c='orange', legend=False, lw=line_width, ylim=(lower, higher));
mx3 = msft.plot(x='qtr', y='eps_est', ax=ax21, c='green', legend=False, lw=line_width, ylim=(lower, higher));
ix3 = ibm.plot(x='qtr', y='eps_est', ax=ax22, c='maroon', legend=False, lw=line_width, ylim=(lower, higher));
ox3 = orcl.plot(x='qtr', y='eps_est', ax=ax31, c='black', legend=False, lw=line_width, ylim=(lower, higher));

ax3.set_ylabel('Consensus Earnings Estimates', fontsize=12, fontweight='bold')
ax3.set_xlabel('Quarter', fontsize=12, fontweight='bold')
ax3.set_title('AAPL',color='blue', fontsize=17, fontweight='bold')

bx3.set_ylabel('Consensus Earnings Estimates', fontsize=12, fontweight='bold')
bx3.set_xlabel('Quarter', fontsize=12, fontweight='bold')
bx3.set_title('BMRC',color='Orange', fontsize=17, fontweight='bold')

mx3.set_ylabel('Consensus Earnings Estimates', fontsize=12, fontweight='bold')
mx3.set_xlabel('Quarter', fontsize=12, fontweight='bold')
mx3.set_title('MSFT',color='green', fontsize=17, fontweight='bold')

ix3.set_ylabel('Consensus Earnings Estimates', fontsize=12, fontweight='bold')
ix3.set_xlabel('Quarter', fontsize=12, fontweight='bold')
ix3.set_title('IBM',color='maroon', fontsize=17, fontweight='bold')

ox3.set_ylabel('Consensus Earnings Estimates', fontsize=12, fontweight='bold')
ox3.set_xlabel('Quarter', fontsize=12, fontweight='bold')
ox3.set_title('ORCL',color='black', fontsize=17, fontweight='bold')


for i in index:
    # ax3.text(i, aapl['eps_est'][i], aapl['eps_est'][i], color='blue', fontsize=12, fontweight='bold')
    # bx3.text(i, bmrc['eps_est'][i], bmrc['eps_est'][i], color='orange', fontsize=12, fontweight='bold')
    # mx3.text(i, msft['eps_est'][i], msft['eps_est'][i], color='green', fontsize=12, fontweight='bold')
    # ix3.text(i, ibm['eps_est'][i], ibm['eps_est'][i], color='maroon', fontsize=12, fontweight='bold')
    print(aapl['eps_est'][i], bmrc['eps_est'][i], msft['eps_est'][i], ibm['eps_est'][i])
for i in index[:2]:
    #ox3.text(i, orcl['eps_est'][i], orcl['eps_est'][i], color='black', fontsize=12, fontweight='bold')
    print(orcl['eps_est'][i])

```
