```python
import matplotlib.pyplot as plt
import yfinance as yf

%config InlineBackend.figure_format = 'svg'
```

```python
tickers = ['MSFT', 'ORCL', 'IBM', 'AAPL', 'BMRC']
start_date = '2018-10-01'
end_date = '2018-12-31'
data = yf.download(tickers = tickers, start = start_date, end = end_date)
```

```python
data = yf.download(tickers, start_date, end_date)
data.head()
```

```python
data.columns
```

```python
df = data['Adj Close']
df.head()
```

```python
df.plot()
```

```python
w = [32, 100, 150,350, 210]
res_data = df.pct_change()[1:]
res_data
```

```python
weighted_returns = (w * res_data)
weighted_returns
```

```python
portf_returns = weighted_returns.sum(axis=1)
portf_returns
```

```python
fig = plt.figure()
ax1 = fig.add_axes([0.1,0.1,0.8,0.8])
ax1.hist(portf_returns, bins = 60)
ax1.set_xlabel('Portfolio returns')
ax1.set_ylabel("Freq")
ax1.set_title("Portfolio Returns by days")
plt.show(); 
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('/content/file_destination/stocks_latest/earnings_latest.csv')
#df[df['symbol'] == 'AAPL'][(df['date']<='2019-01-01') & ('2018-10-01'<=df['date'])]
new_df = df[df['symbol'].isin(['AAPL','IBM','MSFT','ORCL','BMRC'])][df['qtr'].isin(['04/2019','03/2019','02/2019','01/2019','12/2018','11/2018','10/2018','09/2018','08/2018'])]
new_df
```

```python
prep = new_df[['symbol', 'qtr','eps_est']]
prep
plt.style.use('seaborn-whitegrid')
#prep = prep.set_index('qtr')
aapl,bmrc,msft,ibm,orcl = prep[prep['symbol']=='AAPL'],prep[prep['symbol']=='BMRC'],prep[prep['symbol']=='IBM'],prep[prep['symbol']=='MSFT'],prep[prep['symbol']=='ORCL']
aapl.plot('qtr', 'eps_est');
bmrc.plot('qtr', 'eps_est');
msft.plot('qtr', 'eps_est');
ibm.plot('qtr', 'eps_est');
orcl.plot('qtr', 'eps_est');
fig, axes = plt.subplots(nrows=3, ncols=2, figsize=(12,8))
ax11 = axes[0][0]
ax12 = axes[0][1]
ax21 = axes[1][0]
ax22 = axes[1][1]
ax31 = axes[2][0]
line_width = 2.5
lower = prep['eps_est'].min() - 0.5

higher = prep['eps_est'].max() + 0.5
aapl.plot(x='qtr', y='eps_est', ax=ax11, c='blue', legend=False, lw=line_width, ylim=(lower, higher));
bmrc.plot(x='qtr', y='eps_est', ax=ax12, c='orange', legend=False, lw=line_width, ylim=(lower, higher));
msft.plot(x='qtr', y='eps_est', ax=ax21, c='green', legend=False, lw=line_width, ylim=(lower, higher));
ibm.plot(x='qtr', y='eps_est', ax=ax22, c='brown', legend=False, lw=line_width, ylim=(lower, higher));
orcl.plot(x='qtr', y='eps_est', ax=ax31, c='grey', legend=False, lw=line_width, ylim=(lower, higher));
```
